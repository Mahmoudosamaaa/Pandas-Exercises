### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
path = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv"
titanic = pd.read_csv(path)
titanic
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
plt.pie(titanic['Sex'].value_counts().tolist(),labels=['male', 'female'],autopct='%1.2f%%')
plt.title("Males and Females Proportion")
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
grouped = titanic.groupby('Sex')
for gender, group_data in grouped:
    plt.scatter(group_data['Age'], group_data['Fare'], label=gender)
#plt.scatter(titanic['Age'],titanic['Fare'], label = titanic['Sex'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Relation between Fare and Age')
plt.show()
```

### Step 7. How many people survived?

```python
titanic['Survived'].sum()
```

```python
survived_notsurvived = titanic['Survived'].value_counts().tolist()
survived_label = ['Not Survived', 'Survived']

plt.pie(survived_notsurvived, labels=survived_label, autopct='%1.2f%%')
plt.title("Survived and not Survived people")
plt.show()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.hist(titanic['Fare'])
plt.xlabel('Fare payed')
plt.ylabel('frequncy')
plt.title('Fare payed Histogram')
plt.show()
```
